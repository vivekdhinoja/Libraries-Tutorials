# Matplotlib: -

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

## Functional Method: -

```python
plt.plot(x, y)
plt.title('Title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Object Oriented Method: -

```python
fig = plt.figure()
axe = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe.plot(x, y)
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig1.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x, y)
axes2.plot(y, x)
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```

```python
fig2, axes = plt.subplots(nrows=1, ncols=2)

for curr_axe in axes:
    curr_axe.plot(x, y)
    
axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')
```

## Figure Size and DPI: -

```python
fig = plt.figure(figsize=(3, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig.savefig('my_fig_tut_pic.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='X Square')
ax.plot(x, x ** 3, label='X Cube')
ax.legend()
```

## Plot Appearience: -

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='blue', lw=2, alpha=0.5, ls='--', marker='o', markersize=20, 
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
